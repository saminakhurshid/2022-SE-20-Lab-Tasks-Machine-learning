```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

#Load the dataset

```python
data = pd.read_csv('/content/Dataset  (1).csv')
data
```

#visual style for plots

```python
sns.set(style="whitegrid")
```

#Top 10 countries with the most restaurants in the dataset

```python
top_countries = data['Country Code'].value_counts().head(10)
plt.figure(figsize=(8, 5))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="viridis")
plt.title("Top 10 Countries with the Most Restaurants")
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.show()
```

#Distribution of aggregate ratings

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Aggregate rating'], bins=20, kde=True, color='blue')
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

#Top 10 cuisines served across restaurants

```python
cuisine_count = data['Cuisines'].value_counts().head(10)
plt.figure(figsize=(8, 5))
sns.barplot(y=cuisine_count.index, x=cuisine_count.values, palette="plasma")
plt.title("Top 10 Cuisines Served")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Type")
plt.show()
```

# Average cost for two vs. Aggregate rating

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Aggregate rating', y='Average Cost for two', hue='Rating text', alpha=0.7)
plt.title("Average Cost for Two vs. Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Average Cost for Two")
plt.show()
```

#Table booking availability by rating

```python
table_booking = data['Has Table booking'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=table_booking.index, y=table_booking.values, palette="Set2")
plt.title("Table Booking Availability")
plt.xlabel("Has Table Booking")
plt.ylabel("Number of Restaurants")
plt.show()
```

#Online delivery availability

```python
online_delivery = data['Has Online delivery'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=online_delivery.index, y=online_delivery.values, palette="coolwarm")
plt.title("Online Delivery Availability")
plt.xlabel("Has Online Delivery")
plt.ylabel("Number of Restaurants")
plt.show()
```

#Price range distribution

```python
price_range = data['Price range'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
sns.barplot(x=price_range.index, y=price_range.values, palette="viridis")
plt.title("Price Range Distribution")
plt.xlabel("Price Range (1 to 4)")
plt.ylabel("Number of Restaurants")
plt.show()
```

#Top 10 cities with the most restaurants

```python
top_cities = data['City'].value_counts().head(10)
plt.figure(figsize=(8, 5))
sns.barplot(y=top_cities.index, x=top_cities.values, palette="magma")
plt.title("Top 10 Cities with the Most Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.show()
```

#Distribution of votes across restaurants

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Votes'], bins=50, kde=True, color='green')
plt.title("Distribution of Votes Across Restaurants")
plt.xlabel("Number of Votes")
plt.ylabel("Frequency")
plt.show()
```

#Rating text distribution

```python
rating_text_count = data['Rating text'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=rating_text_count.index, y=rating_text_count.values, palette="pastel")
plt.title("Distribution of Rating Text")
plt.xlabel("Rating Text")
plt.ylabel("Number of Restaurants")
plt.show()
```
