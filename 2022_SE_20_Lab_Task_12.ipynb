```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
```

#Loading the Dataset

```python
data = pd.read_excel('/content/Online Retail.xlsx')
data
```

```python
data.head()
```

#Data Preprocessing

```python
# Drop rows with missing values
data = data.dropna()

# Convert 'InvoiceDate' to datetime format
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Remove rows with negative or zero quantities and unit prices
data = data[(data['Quantity'] > 0) & (data['UnitPrice'] > 0)]

# Create a 'TotalPrice' column
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']

# Grouping by 'CustomerID' to calculate RFM (Recency, Frequency, Monetary)
rfm = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (data['InvoiceDate'].max() - x.max()).days,  # Recency
    'InvoiceNo': 'count',  # Frequency
    'TotalPrice': 'sum'     # Monetary
}).rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'TotalPrice': 'Monetary'})

("\nRFM Table:")
(rfm.head())
```

#Data Normalization

```python
# Normalize the RFM values using StandardScaler
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)

print("\nScaled RFM Values:")
print(rfm_scaled[:5])
```

#Finding Optimal K using Elbow Method

```python
# Elbow method to determine the optimal number of clusters
sse = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    sse.append(kmeans.inertia_)

# Plotting the Elbow curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

#Applying K-Means Clustering

```python
# Choose the optimal K (based on the Elbow plot)
optimal_k = 4  # Adjust this based on the Elbow plot

# Applying KMeans with the optimal K
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
rfm['KMeans_Cluster'] = kmeans.fit_predict(rfm_scaled)

# Silhouette Score for KMeans
kmeans_silhouette = silhouette_score(rfm_scaled, rfm['KMeans_Cluster'])
print(f"K-Means Silhouette Score: {kmeans_silhouette:.2f}")
```

#Applying DBSCAN Clustering

```python
# Applying DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
rfm['DBSCAN_Cluster'] = dbscan.fit_predict(rfm_scaled)

# Silhouette Score for DBSCAN (ignoring noise points)
dbscan_labels = rfm['DBSCAN_Cluster']
dbscan_silhouette = silhouette_score(rfm_scaled[dbscan_labels != -1], dbscan_labels[dbscan_labels != -1])
print(f"DBSCAN Silhouette Score: {dbscan_silhouette:.2f}")
```

#Plotting and Comparing Clusters

```python
# Plotting K-Means Clusters
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(rfm_scaled[:, 0], rfm_scaled[:, 1], c=rfm['KMeans_Cluster'], cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Recency')
plt.ylabel('Frequency')

# Plotting DBSCAN Clusters
plt.subplot(1, 2, 2)
plt.scatter(rfm_scaled[:, 0], rfm_scaled[:, 1], c=rfm['DBSCAN_Cluster'], cmap='plasma')
plt.title('DBSCAN Clustering')
plt.xlabel('Recency')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Noise Detection with DBSCAN

```python
# Count of noise points in DBSCAN (-1 represents noise)
noise_points = (rfm['DBSCAN_Cluster'] == -1).sum()
print(f"Number of Noise Points Detected by DBSCAN: {noise_points}")
```
